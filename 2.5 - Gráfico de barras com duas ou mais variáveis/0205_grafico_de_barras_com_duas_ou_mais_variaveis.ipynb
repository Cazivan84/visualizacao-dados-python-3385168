# Gráfico de barras com duas ou mais variáveis

Aprenda a construir gráfico de barras agrupadas e empilhadas utilizando Matplotlib.


Utilizando um conjunto de dados, modificados para fins didáticos, sobre filmes indicados e ganhadores do Oscar em 2022.

Fonte original dos dados: [Kaggle](https://www.kaggle.com/datasets/pushpakhinglaspure/oscar-dataset?resource=download)


```python
# Importa bibliotecas

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Realiza leitura dos dados

df = pd.read_csv('/content/0204_grafico_de_barras_com_duas_ou_mais_variaveis.csv')
```

```python
df.sample(5)
```

```python
# Ordena em ordem decrescente a coluna 'Nomination'

df.sort_values(by='Nomination', ascending=True, inplace=True)
```

```python
# Imprime na tela as cinco primeiras linhas

df.head()
```

```python
# Cria lista com os títulos dos filmes

filmes = df['Film'].to_list()
```

```python
# Cria lista com total de indicações

total_indicacao = df['Nomination'].to_list()
```

```python
# Cria lista com total de prêmios

total_premio = df['Award'].to_list()
```

# Gráfico de barras empilhadas

```python
fig, ax = plt.subplots()

bar_container_indicacao = ax.bar(filmes, total_indicacao, color='purple')
bar_container_premiacao = ax.bar(filmes, total_premio, bottom=total_indicacao, color='green')
ax.bar_label(bar_container_indicacao, label_type='center')
ax.bar_label(bar_container_premiacao, label_type='center')
plt.xticks(rotation = 90)

ax.set_ylabel('Total de indicações')
ax.set_title('Indicados para o Oscar 2022')

plt.show()
```

# Gráfico de barras agrupadas

```python
# Criando visualização a partir do Pandas

df.plot(
    x='Film',
    y=['Award', 'Nomination'],
    title='Oscar 2022',
    ylabel='Quantidade',
    color=['olive', 'purple'],
    kind='bar')

plt.show()
```
