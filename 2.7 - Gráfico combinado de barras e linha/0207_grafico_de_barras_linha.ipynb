# Gráfico de barras e linha
Aprenda a construir gráficos de barras e linha com Matplotlib. Utilizando dados fictícios do faturamento e metas de uma empresa.

```python
# Importa bibliotecas

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importa os dados para dataframe

df = pd.read_csv('/content/0207_grafico_de_barras_linha.csv')
```

```python
# Imprime as cinco primeiras linhas do dataframe

df.head()
```

```python
# Cria lista com a coluna "faturamento"

faturamento = df['faturamento'].to_list()
```

```python
# Cria lista da coluna "meta"

meta = df['meta'].to_list()
```

```python
# Cria lista da coluna "mes"

mes = df['mes'].to_list()
```

```python
# Cria visualização de barras

fig, ax = plt.subplots()

ax.bar(mes, meta, color='gray')
ax.set(ylabel='Valor em R$')
ax.set_title('Acompanhamento de metas - 2023', pad=20)

plt.xticks(rotation = 90)

plt.show()
```

```python
# Cria visualização de linha

fig, ax = plt.subplots()

ax.plot(mes, faturamento, color='purple', linewidth=2.5, label='faturamento')
ax.set(ylabel='Valor em R$')
ax.set_title('Acompanhamento faturamento - 2023', pad=20)

plt.xticks(rotation = 90)

ax.legend()
plt.show()
```

```python
# Cria visualização combinada

fig, ax = plt.subplots()

ax.bar(mes, meta, color='gray', label='meta')
ax.plot(mes, faturamento, color='purple', linewidth=2.5, label='faturamento')
ax_faturamento = ax.secondary_yaxis('right')
ax.set(ylabel='Valor em R$')
ax.set_title('Acompanhamento de metas e faturamento - 2023', pad=20)

plt.xticks(rotation = 90)

ax.legend()
plt.show()
```
