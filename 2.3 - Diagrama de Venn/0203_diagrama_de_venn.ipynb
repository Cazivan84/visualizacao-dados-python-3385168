```python
# Importa as bibliotecas

import matplotlib.pyplot as plt

from matplotlib_venn import venn2, venn3, venn3_circles
```

```python
# Declara as variáveis com quantitativos de registros

bd_apenas_antiga = 800
bd_antiga_intersecacao_atual = 400
bd_apenas_atual = 2100
bd_nova_contida_atual = 200
```

## Comparando apenas a base de dados antiga com a atual:

```python
# Cria visualização com dois conjuntos

venn2(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual), # Sintaxe: (A, B, C)
    set_labels=("BD antiga", "BD atual")
)

plt.show()
```

## Comparando as base de dados antiga, atual e a nova:

```python
# Cria visualização co três conjuntos

venn3(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0), # Sintaxe: (A, B, C, D, E, F, G)
    set_labels=("BD antiga", "BD atual", "BD nova")
)

plt.title('Comparação entre bases de dados: duplicação de registros')

plt.show()
```

## Customizações

## Cor

```python
# Adiciona parâmetro "set_colors" para alterar a cor padrão

venn3(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0),
    set_labels=("BD antiga", "BD atual", "BD nova"),
    set_colors=("purple", "skyblue", "yellow")
)

plt.show()
```

### Contorno

```python
# Cria contorno para os conjuntos

venn3(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0),
    set_labels=("BD antiga", "BD atual", "BD nova"),
    set_colors=("purple", "skyblue", "yellow")
)

venn3_circles(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0),
    linewidth=0.6
)

plt.show()
```

```python
# Customiza linha do contorno

venn3(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0),
    set_labels=("BD antiga", "BD atual", "BD nova"),
    set_colors=("purple", "skyblue", "yellow")
)

venn3_circles(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0),
    linewidth=0.6,
    linestyle="dashed"
)

plt.show()
```

```python
# Customiza cor da linha do contorno

venn3(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0),
    set_labels=("BD antiga", "BD atual", "BD nova"),
    set_colors=("purple", "skyblue", "yellow")
)

venn3_circles(
    subsets=(bd_apenas_antiga, bd_apenas_atual, bd_antiga_intersecacao_atual, 0, 0, bd_nova_contida_atual, 0),
    linewidth=0.6,
    linestyle="dashed",
    color="green"
)

plt.show()
```
