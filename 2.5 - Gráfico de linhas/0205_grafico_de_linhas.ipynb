# Gráfico de linhas
Aprenda a construir gráficos de linhas usando Matplotlib.

Utilizando um conjunto de dados, modificados para fins didáticos, do Observatório da Criança e do Adolescente sobre taxa bruta e líquida de matrículas na pré-escola entre 2010 e 2019.

Fonte original dos dados: [Base dos dados](https://basedosdados.org/dataset/11b867be-e626-4da7-9dad-f5dd930c67fe?table=687a48b3-8e09-46da-80e0-b734ea0b51ec)


```python
# Importa bibliotecas

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importa os dados para dataframe

df = pd.read_csv('/content/0205_grafico_de_linhas.csv')
```

```python
# Imprime as cinco primeiras linhas do dataframe

df.head()
```

## Observando oscilação da taxa apenas do município de Recife

```python
# Cria lista com as taxas apenas de Recife

tx_recife = df[df['municipio'] == 'Recife']['taxa_liquida_matricula_pre_escola'].to_list()
```

```python
# Cria lista com anos

anos = df[df['municipio'] == 'Recife']['ano'].to_list()
```

```python
# Cria visualização

fig, ax = plt.subplots()
ax.plot(anos, tx_recife)

plt.show()
```

## Customização

### Etiquetas

```python
fig, ax = plt.subplots()
ax.plot(anos, tx_recife)

ax.set(xlabel='Anos', ylabel='Taxa') # Insere etiquetas
ax.set_title('Taxa de escolarização líquida da pré-escola 2010 e 2019 em Recife', pad=20) # Insere título

plt.show()
```

### Cor

```python
fig, ax = plt.subplots()
ax.plot(anos, tx_recife, color='purple') # Insere cor

ax.set(xlabel='Anos', ylabel='Taxa')
ax.set_title('Taxa de escolarização líquida da pré-escola 2010 e 2019 em Recife', pad=20)

plt.show()
```

### Linha: espessura e estilo

```python
fig, ax = plt.subplots()
ax.plot(anos, tx_recife, color='purple', linewidth=3, linestyle='--') # Insere espessura e estilo

ax.set(xlabel='Anos', ylabel='Taxa')
ax.set_title('Taxa de escolarização líquida da pré-escola 2010 e 2019 em Recife', pad=20)

plt.show()
```

## Gráfico de linha com marcação

```python
fig, ax = plt.subplots()
ax.plot(anos, tx_recife, color='purple', marker='o') # Insere cor

ax.set(xlabel='Anos', ylabel='Taxa')
ax.set_title('Taxa de escolarização líquida da pré-escola 2010 e 2019 em Recife', pad=20)

plt.show()
```

## Gráfico comparativo entre as taxas de escolarização das capitais do Nordeste

```python
municipios = df['municipio'].unique()
```

```python
variaveis_municipios = ['sao_luis', 'teresina', 'fortaleza', 'natal', 'joao_pessoa', 'recife', 'maceio', 'aracaju', 'salvador']
```

```python
ano = df['ano'].unique()
```

```python
fig, ax = plt.subplots(figsize=(15, 7), layout='constrained')

for variavel, municipio in zip(variaveis_municipios, municipios):
  globals()[variavel] = df[df['municipio'] == municipio]['taxa_liquida_matricula_pre_escola'].to_list()
  ax.plot(ano, globals()[variavel], linewidth=2.2, marker='D', label=municipio)

ax.set(xlabel='Anos', ylabel='Taxa')
ax.set_title('Taxa de escolarização líquida da pré-escola 2010 e 2019 nos estados do Nordeste', pad=20)
ax.legend()

plt.savefig('taxa_escolarizacao_liquida_estados_nordeste.png')
```
