# Histograma
Aprenda a construir histograma com Matplotlib e Seaborn. Utilizando um conjunto de dados, modificados para fins didáticos, sobre características de flores.


Fonte original dos dados: [Kaggle](https://www.kaggle.com/datasets/saurabh00007/iriscsv)


```python
# Importa bibliotecas

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/0208_histograma.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df, x="SepalWidthCm")

plt.show()
```

## Customização

## Título

```python
plt.figure(figsize=(7, 4))
plt.title('Iris - largura da sépala')

sns.histplot(data=df, x="SepalWidthCm")

plt.show()
```

### Etiqueta dos eixos

```python
plt.figure(figsize=(7, 4))
plt.title('Iris - largura da sépala')
plt.xlabel('Largura da sépala')
plt.ylabel('Frequência')

sns.histplot(data=df, x="SepalWidthCm")

plt.show()
```

### Cor

```python
plt.figure(figsize=(7, 4))
plt.title('Iris - largura da sépala')
plt.xlabel('Largura da sépala')
plt.ylabel('Frequência')

sns.histplot(data=df, x="SepalWidthCm", color='green')

plt.show()
```

### Estimativa de densidade de Kernel

```python
plt.figure(figsize=(7, 4))
plt.title('Iris - largura da sépala')
plt.xlabel('Largura da sépala')
plt.ylabel('Frequência')

sns.histplot(data=df, x="SepalWidthCm", kde=True, color='green')

plt.show()
```
