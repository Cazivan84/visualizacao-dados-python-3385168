# Box plot
Aprenda a construir box plot com Matplotlib e Seaborn. Utilizando um conjunto de dados, modificados para fins didáticos, sobre passageiros do Titanic.


Fonte original dos dados: [Kaggle](https://www.kaggle.com/datasets/yasserh/titanic-dataset)


```python
# Importa bibliotecas

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/0210_boxplot.csv')
```

```python
df.head()
```

## Distribuição da tarifa dos bilhetes

```python
plt.figure(figsize=(20, 4))
sns.boxplot(data=df,
            x='Fare')

plt.show()
```

## Distribuição da tarifa dos bilhetes por gênero

```python
plt.figure(figsize=(20, 4))
sns.boxplot(data=df,
            x='Fare',
            y='Sex',
            hue='Sex')

plt.show()
```

## Customização

### Título e etiquetas

```python
plt.figure(figsize=(20, 4))
sns.boxplot(data=df,
            x='Fare',
            y='Sex',
            hue='Sex')

plt.title('Tarifa do bilhete por gênero') # Insere título
plt.xlabel('Tarifa') # Insere etiqueta do eixo x
plt.ylabel('Gênero') # Insere etiqueta do eixo y

plt.show()
```

### Cor

```python
plt.figure(figsize=(20, 4))
sns.boxplot(data=df,
            x='Fare',
            y='Sex',
            hue='Sex',
            palette='Set2') # Modifica cores

plt.title('Tarifa do bilhete por gênero')
plt.xlabel('Tarifa')
plt.ylabel('Gênero')

plt.show()
```

### Espessura da linha

```python
plt.figure(figsize=(20, 4))
sns.boxplot(data=df,
            x='Fare',
            y='Sex',
            hue='Sex',
            palette='Set2',
            linewidth=3) # Define espessura da linha

plt.title('Tarifa do bilhete por gênero')
plt.xlabel('Tarifa')
plt.ylabel('Gênero')

plt.show()
```

### Gradação

```python
fig, ax = plt.subplots(figsize=(20,4))

sns.boxplot(data=df,
            x='Fare',
            y='Sex',
            hue='Sex',
            palette='Set2',
            linewidth=3)

ax.xaxis.set_ticks(list(range(0,550,25))) # Estabelece grade com intervalo múltiplo de 25

plt.title('Tarifa do bilhete por gênero')
plt.xlabel('Tarifa')
plt.ylabel('Gênero')

plt.show()
```
