# Diagrama de dispersão
Aprenda a construir diagrama de dispersão com Matplotlib e Seaborn. Utilizando um conjunto de dados, modificados para fins didáticos, sobre passageiros do Titanic.


Fonte original dos dados: [Kaggle](https://www.kaggle.com/datasets/yasserh/titanic-dataset)

```python
# Importa bibliotecas

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/0211_diagrama_de_dispersao.csv')
```

```python
# Cria nova coluna com categoria a partir da coluna 'Survived'

df['Estado'] = df.apply(lambda row: 'sobrevivente' if row['Survived'] == 1 else 'não sobrevivente', axis=1)
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(20,4))

sns.scatterplot(data=df,
                x='Fare',
                y='Age',
                hue='Sex',
                palette='husl')

ax.xaxis.set_ticks(list(range(0,550,20))) # Estabelece grade com intervalo múltiplo de 20

plt.title('Tarifa do bilhete por gênero e idade')
plt.xlabel('Tarifa')
plt.ylabel('Idade')

plt.show()
```

## Sobreviventes por sexo, idade e preço de bilhete

```python
graficos = sns.relplot(data=df,
                      x='Fare',
                      y='Age',
                      col='Sex',
                      hue='Estado',
                      style='Estado',
                      markers={'sobrevivente':"o", 'não sobrevivente':"P"},
                      palette='deep',
                      kind='scatter')

sns.move_legend(graficos, 'upper right', frameon=True)

plt.show()
```
